```python
import camelot
import pandas as pd
from PyPDF2 import PdfReader

pdf_path = "/content/BESTCOMPANYTY24.pdf"
reader = PdfReader(pdf_path)
num_pages = len(reader.pages)
print(f"\n Total pages detected: {num_pages}")
all_tables = []

for page_num in range(1, num_pages + 1):
    tables = camelot.read_pdf(pdf_path, pages=str(page_num), flavor='lattice')
    if len(tables) == 0:
        print(" No tables found with 'lattice', switching to 'stream' mode.")
        tables = camelot.read_pdf(pdf_path, pages=str(page_num), flavor='stream')

    for table in tables:
        df = table.df
        all_tables.append(df)

```

```python
for i in range(len(all_tables)):
   all_tables[i].to_csv(f"{i}.csv", index=False)
```

# PDFPlumber

```python
!pip install -qU pdfplumber
```

```python
import pdfplumber #No
import pandas as pd

pdf_path = "/content/Statement20230930 (3).pdf"

table_settings = {
    "vertical_strategy": "text",
    "horizontal_strategy": "text",
    "snap_tolerance": 3,
    "join_tolerance": 8,
    "edge_min_length": 3,
    "min_words_vertical": 10,
    "min_words_horizontal": 10,
    "text_tolerance": 12,  # tweak this value if columns merge/split wrongly
}

all_tables=[]

with pdfplumber.open(pdf_path) as pdf:
    for page_num, page in enumerate(pdf.pages, start=1):
        table = page.extract_table(table_settings)
        if table:
            df = pd.DataFrame(table)
            print(f"Table found on page {page_num}")
            all_tables.append(df)
        else:
            print(f"No table found on page {page_num}")
```

```python
from img2table.document import PDF
from img2table.ocr import TesseractOCR

# Instantiation of the pdf
pdf = PDF(src="/content/page_1.pdf")

# Instantiation of the OCR, Tesseract, which requires prior installation
ocr = TesseractOCR(lang="eng")

# Table identification and extraction
pdf_tables = pdf.extract_tables(ocr=ocr)

# We can also create an excel file with the tables
pdf.to_xlsx('tables.csv',
            ocr=ocr)
```
