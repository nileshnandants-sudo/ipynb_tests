```python
from PIL import Image
import torch
from torch import autocast
from transformers import AutoModelForCausalLM, AutoProcessor

DEVICE = "cuda" if torch.cuda.is_available() else "cpu"
CHOSEN_TASK="ocr"
PROMPTS={"ocr": "OCR"}
```

```python
model_path = "PaddlePaddle/PaddleOCR-VL"
image_path = "/content/batch1-0495.jpg"
```

```python
image=Image.open(image_path).convert("RGB")
image = image.resize((1024, 1024))
```

```python
model = AutoModelForCausalLM.from_pretrained(model_path,
                                             trust_remote_code=True,
                                             torch_dtype = torch.float16,
                                             low_cpu_mem_usage=True,
                                             device_map="auto").eval()
processor = AutoProcessor.from_pretrained(model_path,
                                          trust_remote_code=True)
```

```python
print(model)
```

```python
messages=[{
    "role": "user",
    "content": PROMPTS[CHOSEN_TASK]
}]

text = processor.apply_chat_template(messages, tokenize=False, add_generation_prompt=True)
inputs = processor(text=[text], images=[image], return_tensors="pt")
inputs = {k:(v.to(DEVICE) if isinstance(v, torch.Tensor) else v) for k,v in inputs.items()}

with torch.inference_mode(), autocast(device_type="cuda", dtype=torch.float16):
    generated = model.generate(**inputs,
                               max_new_tokens=256,
                               do_sample=False,
                               use_cache=True)

resp = processor.batch_decode(generated, skip_special_tokens=True)[0]
answer = resp.split(text)[-1].strip()
print(answer)
```
